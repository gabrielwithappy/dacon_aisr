## Inference

```python
import subprocess
from datetime import datetime
from pytz import timezone
import os

def inference_for_ensemple(model_pth, dst_ensemble_folder):
  """
  <Folder Structure>
  /Real-ESRGAN
    ㄴ inputs
  /config_backup
  /datasets
  /utility
    ㄴ ensemble
      ㄴ model_1
      ㄴ model_2
      ㄴ model_3
      ㄴ model_4
      ㄴ result (the ensemble images of models )    
  inference by using model pth and move images to each folder for ensemble
      Args:
          model_pth : the name of the model-pth saved in the experiments/tags/ folder
          dst_ensemble_foler : a tuple which contain (folder name, model description)
                               folder name should be the same name of foler in /utility/ensemble/
      Retruns:
          None
  """
  

  for idx in range(len(model_pth)):
    arg_model_path = f'tags/{model_pth[idx]}'
    command = f"python3 inference_realesrgan.py --model_path " + arg_model_path +  " --suffix ''"



    try:
      os.system(command)

      dst_path = f"{ROOT}/../utility/ensemble/{dst_ensemble_folder[idx]}"
      if not os.path.isdir(dst_path):
        os.mkdir(dst_path)

      if os.path.exists(f"dst_path"):
        shutil.rmtree(f"dst_path")
      shutil.copytree(src=f'{ROOT}/results/',
                      dst=f"dst_path")
      print(f"copyed results to ensemble folder {dst_ensemble_folder[idx]}")
      #shutil.rmtree(f'{ROOT}/results/')
    except:
      print("ensemble error!")
    

  cur_date = datetime.now(timezone('Asia/Seoul')).strftime('%Y-%m-%d_%H:%M:%S')
  timestamp_file = f"{ROOT}/../utility/ensemble/ensemble_model_info_{cur_date}.log"

  with open(timestamp_file, "w") as f:
    f.write("Ensemble Info ===\n")
    for idx in range(len(model_pth)):
      f.write(f"copyed {model_pth[idx]} results to {dst_ensemble_folder[idx]}\n")
    f.write(f"saved {cur_date}")
    f.close()
    print(f"saved {timestamp_file}")
```
